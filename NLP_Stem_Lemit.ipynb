```python
from nltk.tokenize import sent_tokenize,word_tokenize
```

```python
nltk.download('punkt_tab')
sentences=sent_tokenize(text)
```

Word Tokenizer


```python
Words=word_tokenize(text)
```

Lower

```python
lower_sent=[sentence.lower() for sentence in sentences]
```

```python
lower_sent
```

```python
word_low =[word.lower() for word in Words]
```

```python
len(word_low)
```

```python
from  nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
stop_words=set(stopwords.words('english'))
```

```python
clean_data=[word for word in word_low if word not in stop_words and word.isalpha()]
```

```python
len(clean_data)
```

Stemming

```python
from nltk.stem import PorterStemmer,WordNetLemmatizer
```

```python
stemmer=PorterStemmer()
```

```python
stemmed_data = [stemmer.stem(word) for word in clean_data]
stemmed_data
```

Lemitization

```python
lematizer=WordNetLemmatizer()
nltk.download('wordnet')
```

```python
lematize_data=[lematizer.lemmatize(word) for word in clean_data]
```

```python
lematize_data
```

UniGram

```python
from nltk import ngrams
```

```python
uni_grams=list(ngrams(clean_data,1))
```

```python
uni_grams
```

```python
bi_grams=list(ngrams(clean_data,2))
bi_grams
```
